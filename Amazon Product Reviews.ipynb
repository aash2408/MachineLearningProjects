```python
import nltk
#nltk.download("punkt")
from nltk.tokenize import word_tokenize
```

```python
df_reviews['tokenized'] = df_reviews['no_contractions_review'].apply(word_tokenize)
```

```python
df_reviews
```

```python
df_reviews['tokens_lowercase'] = df_reviews['tokenized'].apply(lambda token_list:[word.lower() for word in token_list])
```

```python
df_reviews['tokens_lowercase']
```

```python
import string
punc = string.punctuation
```

```python
#removing punctuation
df_reviews['tokens_no_punc'] = df_reviews['tokens_lowercase'].apply(lambda token_list:[word for word in token_list if word not in punc])
df_reviews['tokens_no_punc']
```

```python
#Stopwords
#nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
df_reviews['tokens_no_stopwords'] = df_reviews['tokens_no_punc'].apply(lambda token_list:[word for word in token_list if word not in stop_words])
```

```python
df_reviews['tokens_no_stopwords']
```

```python
#parts of speech tagging
#nltk.download('averaged_perceptron_tagger')
```

```python
df_reviews['pos_tags'] = df_reviews['tokens_no_stopwords'].apply(nltk.tag.pos_tag)
df_reviews['pos_tags']
```

```python
#nltk.download('wordnet')
from nltk.corpus import wordnet
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
df_reviews['wordnet_pos_tags'] = df_reviews['pos_tags'].apply(lambda tuple_list:[(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in tuple_list])
```

```python
df_reviews['wordnet_pos_tags']
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
```

```python
df_reviews['lemmatized'] = df_reviews['wordnet_pos_tags'].apply(lambda tuple_list:[wnl.lemmatize(word, net) for (word, net) in tuple_list])
df_reviews['lemmatized']
```

```python
df_reviews['lemmatized_text'] = [' '.join(map(str, word_list)) for word_list in df_reviews['lemmatized']]
```

```python
df_reviews['lemmatized_text']
```

```python
df_reviews
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#TFIDF vectorization
vect = TfidfVectorizer(max_df = 0.60, min_df = 0.06, ngram_range = (1,2), use_idf = True)
```

```python
vect_text = vect.fit_transform(df_reviews['lemmatized_text'])
```

```python
vect.get_feature_names()
```

```python
#creating LDA model
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda_model = LatentDirichletAllocation(n_components = 12, random_state = 10, learning_method = 'online', max_iter = 10)
lda_top = lda_model.fit_transform(vect_text)
```

```python
lda_top
```

```python
lda_top.shape
```

```python
topics = lda_model.components_
topics
```

```python
topics.shape
```

```python
#Creating vocabulary array which will represent the corpus
vocab_tf_idf = vect.get_feature_names()
vocab_tf_idf
```

```python
len(vocab_tf_idf)
```

```python
#Retrieve the topics
n_top_words = 7

for i, topic_dist in enumerate(topics):
    print("Topic " , i , " : " , topic_dist, "\n")
```

```python
for i, topic_dist in enumerate(topics):
    #Sorting the above array
    sort_topic_dist = np.argsort(topic_dist)
    #To view the actual words present in those indexes, we can use the vocab created earlier
    topics = np.array(vocab_tf_idf)[sort_topic_dist]
    #using the sorted indexes we are using the words in the vocabulary
    topics = topics[:-n_top_words:-1]
    print("Topic", str(i+1), topics)
```
