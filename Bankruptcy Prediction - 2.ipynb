```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Bankruptcy prediction.csv')
df
```

```python
#finding out the number of null values
df.isnull().sum()
```

```python
#finding out the numner of bankrupt
df['Bankrupt?'].value_counts()
```

```python
df['Bankrupt?'].value_counts()
```

```python
#dropping null values
df = df.dropna()
df
```

```python
#Specifying X and Y
X = df.drop(['Bankrupt?'], axis = 1)
Y = df['Bankrupt?']
Y
```

```python
df_corr = df.corr()['Bankrupt?'].abs().sort_values(ascending=False)
df_corr
```

```python
#Splitting X and Y into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
X_train_copy = X_train
X_test_copy = X_test
```

```python
#implementing Smote for oversampling minority class
from imblearn.over_sampling import SMOTE
#counter = Counter(Y_train)
#print('Before', counter)
#over sampling the train dataset using SMOTE
smt = SMOTE()
X_train_sm, Y_train_sm = smt.fit_resample(X_train, Y_train)
#counter = Counter(Y_train_sm)
#print('After', counter)
```

```python
len(Y_train_sm)
```

```python
#Standardizing train and test datasets
#enlisting the numerical features
list_of_numerical_columns = df.columns.tolist()
list_of_numerical_columns.remove(' Liability-Assets Flag')
list_of_numerical_columns.remove(' Net Income Flag')
list_of_numerical_columns.remove('Bankrupt?')
len(list_of_numerical_columns)
```

```python
X_train_sm = pd.DataFrame(data = X_train_sm.values, columns=X_train_copy.columns)
X_train_sm
```

```python
#Standardizing train and test datasets
from sklearn.preprocessing import StandardScaler
for i in list_of_numerical_columns:
    # fit on training data column
    scale = StandardScaler().fit(X_train_sm[[i]])
    # transform the training data column
    X_train_sm[i] = scale.transform(X_train_sm[[i]])
    # transform the testing data column
    X_test[i] = scale.transform(X_test[[i]])
```

```python
len(X_train_sm)
```

```python
X_test
```

```python
Y_train_sm = pd.DataFrame(data = Y_train_sm)
Y_train_sm
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gradient_booster = GradientBoostingClassifier(learning_rate=0.1)
gradient_booster.get_params()
```

```python
gradient_booster.fit(X_train_sm,Y_train_sm)
```

```python
Y_GB_predict = gradient_booster.predict(X_test)
Y_GB_predict
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(Y_GB_predict, Y_test)
```

```python
confusion_matrix(Y_GB_predict, Y_test)
```
